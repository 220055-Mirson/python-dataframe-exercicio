```python
import pandas as pd
import numpy as np

data = {
    "nome": ["ana", "bruno", "carlos", "ana", None],
    "idade": [23, None, 35, 23, 29],
    "cidade": ["Maputo", "beira", "MAPUTO", "nampula", None],
    "salario": ["5000", "6000", None, "7000", None]
}

df = pd.DataFrame(data)
df.to_csv("dados.csv", index=False)

print("Dataset original:")
print(df)

print("\nDetecção de valores nulos (True = nulo):")
print(df.isnull())

print("\nQuantidade de valores nulos por coluna:")
print(df.isnull().sum())

df["nome"] = df["nome"].fillna("mirson")
df["idade"] = df["idade"].fillna(21)
df["cidade"] = df["cidade"].fillna("Maxixe")
df["salario"] = df["salario"].fillna("35000")

print("\nDataset após tratamento:")
print(df)

df.to_csv("dados_limpos.csv", index=False)
```
