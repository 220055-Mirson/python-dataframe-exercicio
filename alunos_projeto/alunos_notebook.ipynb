```python
import pandas as pd

dados = {
    "nome": ["Joao", "Manuel", "Ana", "Ronaldo", "Maria"],
    "idade": [17, 16, 15, 19, 17],
    "curso": ["biologia", "direito", "informatica", "desporto", "geografia"],
    "nota": [14, 13, 13, 11, 16]
}

df = pd.DataFrame(dados)
df.to_csv("alunos.csv", index=False)

print(df.head())
print(df.isnull().sum())

df = df.drop_duplicates()
print(df)

df_filtrado = df[df["nota"] > 15]
print(df_filtrado)

df_filtrado.to_csv("alunos_limpos.csv", index=False)
```
