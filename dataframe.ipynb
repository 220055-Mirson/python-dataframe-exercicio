```python
import numpy as np
import pandas as pd

quantidade = 1000

nomes = ["Aluno_" + str(i) for i in range(quantidade)]

cursos = ["Engenharia", "Informática", "Medicina", "Direito", "Arquitetura"]

cursos_aleatorios = np.random.choice(cursos, quantidade)

notas = np.random.uniform(0, 20, quantidade)
notas = notas.round(2)

df = pd.DataFrame({
    "Nome": nomes,
    "Curso": cursos_aleatorios,
    "Nota": notas
})

print(df.head())
```
